```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def insertion(self,data):
        newnode = Node(data)
        newnode.next = self.head
        self.head = newnode
    def display(self):
        current = self.head
        while current!= None:
            print(current.value,end="->")
            current = current.next
        print("None")
ll = LinkedList()
ll.insertion(10)
ll.insertion(20)
ll.display()
```

```python
def insertionend(self,data):
    newnode = Node(data)
    if self.head is None:
        self.head = newnode
    else:
        current = self.head
        while current.next!=None:
            current = current.next
        current.next = newnode
        print("None")
ll = LinkedList()
ll.insertion(9398388836)
ll.insertion(9390134300)
ll.display()
```

```python
def max_meetings(meetings):
    meetings.sort(key=lambda meeting: meeting[1])
    meeting_count = 0
    last_meeting_end = -1
    for start_time, end_time in meetings:
        if start_time >= last_meeting_end:
            meeting_count += 1
            last_meeting_end = end_time
    return meeting_count
T = int(input())
for i in range(T):
    N = int(input())
    meetings = []
    for j in range(N):
        start, end = map(int, input().split())
        meetings.append((start, end))
    print(max_meetings(meetings))
```
